### Get the number of columns and their names

Create a reusable function to display all column names in a DataFrame

```python
def print_column_names(dataframe):

    print('Column names:')
    # get each column from the DataFrame
    # use enumerate() class to get the coresponding index
    for i, column in enumerate(dataframe.columns):

        # print the column with the corresponding index
        print(f'{i + 1}.', column)
```

```python
# print the number of columns in the dataset
print(f'\nNumber of columns: {len(data.columns)}\n')

# print the all column names using the 'print_column_names()' reusable function
print_column_names(data)
```
